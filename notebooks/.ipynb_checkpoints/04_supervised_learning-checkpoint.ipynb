```python
from itertools import cycle
import matplotlib.pyplot as plt
import numpy as np
from dask.array import average
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, \
    RocCurveDisplay, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier
from importlib import import_module
import importnb

with importnb.Notebook():
        # Import the notebook as a module
        dataPreprocessing = import_module("01_data_preprocessing")
        pcaAnalysis = import_module("02_pca_analysis")
        featureSelection = import_module("03_feature_selection")
```

```python
finalDataset = pcaAnalysis.X_reduced[: , featureSelection.rfe.get_support()]


X, X_test, y,  y_test = train_test_split(finalDataset , dataPreprocessing.label ,train_size= 0.8, random_state=42)

label_binarizer = LabelBinarizer().fit(y)
y_onehot_test = label_binarizer.transform(y_test)  # (n_samples, n_classes)

n_classes = len(np.unique(y))
target_names = ["Class 0", "Class 1", "Class 2", "Class 3", "CLass 4"]
fig, ax = plt.subplots(figsize=(6, 6))
colors = cycle(["aqua", "darkorange", "cornflowerblue","green","red"])
```

```python
lr_model = LogisticRegression()
lr_model.fit(X,y)
lr_prediction = lr_model.predict(X_test)
print("---------- Logistic Regression Results ------------")
print("Accuracy Score : ", accuracy_score(y_test,lr_prediction))
print("Precision Score : ", precision_score(y_test,lr_prediction,average="micro"))
print("Recall Score : ", recall_score(y_test,lr_prediction,average="micro"))
print("F1 Score : ", f1_score(y_test,lr_prediction,average="micro"))
print("ROC AUC Score : ", roc_auc_score(y, lr_model.predict_proba(X), multi_class='ovr'))

fig, ax = plt.subplots(figsize=(6, 6))

for class_id, color in zip(range(n_classes), colors):
    RocCurveDisplay.from_predictions(
        y_onehot_test[:, class_id],
        lr_model.predict_proba(X_test)[:, class_id],
        name=f"ROC curve for {target_names[class_id]}",
        ax=ax,
    )

_   = ax.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title="Logistic Regression Receiver Operating Characteristic\nto One-vs-Rest multiclass",
    )
plt.show()
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X,y)
dt_prediction = dt_model.predict(X_test)
print("---------- Decision Tree Results ------------")
print("Accuracy Score : ", accuracy_score(y_test,dt_prediction))
print("Precision Score : ", precision_score(y_test,dt_prediction,average="micro"))
print("Recall Score : ", recall_score(y_test,dt_prediction,average="micro"))
print("F1 Score : ", f1_score(y_test,dt_prediction,average="micro"))
print("ROC AUC Score : ", roc_auc_score(y, dt_model.predict_proba(X), multi_class='ovr'))

fig, ax = plt.subplots(figsize=(6, 6))

for class_id, color in zip(range(n_classes), colors):
    RocCurveDisplay.from_predictions(
        y_onehot_test[:, class_id],
        dt_model.predict_proba(X_test)[:, class_id],
        name=f"ROC curve for {target_names[class_id]}",
        ax=ax,
    )

_   = ax.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title="Decision Tree Receiver Operating Characteristic\nto One-vs-Rest multiclass",
    )
plt.show()
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X,y)
rf_prediction = rf_model.predict(X_test)
print("---------- Random Forest Results ------------")
print("Accuracy Score : ", accuracy_score(y_test,rf_prediction))
print("Precision Score : ", precision_score(y_test,rf_prediction,average="micro"))
print("Recall Score : ", recall_score(y_test,rf_prediction,average="micro"))
print("F1 Score : ", f1_score(y_test,rf_prediction,average="micro"))
print("ROC AUC Score : ", roc_auc_score(y, rf_model.predict_proba(X), multi_class='ovr'))

fig, ax = plt.subplots(figsize=(6, 6))

for class_id, color in zip(range(n_classes), colors):
    RocCurveDisplay.from_predictions(
        y_onehot_test[:, class_id],
        rf_model.predict_proba(X_test)[:, class_id],
        name=f"ROC curve for {target_names[class_id]}",
        ax=ax,
    )

_   = ax.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title="Random Forest Receiver Operating Characteristic\nto One-vs-Rest multiclass",
    )
plt.show()
```

```python
svm_model = SVC(probability=True)
svm_model.fit(X,y)
svm_prediction = svm_model.predict(X_test)
print("---------- Support Vector Machine Results ------------")
print("Accuracy Score : ", accuracy_score(y_test,svm_prediction))
print("Precision Score : ", precision_score(y_test,svm_prediction,average="micro"))
print("Recall Score : ", recall_score(y_test,svm_prediction,average="micro"))
print("F1 Score : ", f1_score(y_test,svm_prediction,average="micro"))
print("ROC AUC Score : ", roc_auc_score(y, svm_model.predict_proba(X), multi_class='ovr'))

fig, ax = plt.subplots(figsize=(6, 6))

for class_id, color in zip(range(n_classes), colors):
    RocCurveDisplay.from_predictions(
        y_onehot_test[:, class_id],
        svm_model.predict_proba(X_test)[:, class_id],
        name=f"ROC curve for {target_names[class_id]}",
        ax=ax,
    )

_   = ax.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title="Support Vector Machine Receiver Operating Characteristic\nto One-vs-Rest multiclass",
    )
plt.show()
```
