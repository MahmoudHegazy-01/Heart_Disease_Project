```python
from importlib import import_module

from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
from importlib import import_module
import importnb

with importnb.Notebook():
        # Import the notebook as a module
        dataPreprocessing = import_module("01_data_preprocessing")
```

```python
pca = PCA().fit(dataPreprocessing.df)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.grid()
plt.show()
```

```python
cov_matrix = np.cov(dataPreprocessing.df.T)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("Eigenvalues : ", eigenvalues)
print("Eigenvectors : ", eigenvectors)
```

```python
X_reduced = PCA(n_components=11).fit_transform(dataPreprocessing.df)



plt.scatter(X_reduced[:,0], X_reduced[:,1])
plt.show()
```

```python
pca = PCA(n_components= 11).fit(dataPreprocessing.df)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.grid()
plt.show()
```
