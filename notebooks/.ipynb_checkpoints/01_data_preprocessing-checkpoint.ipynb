```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../data/heart_disease.csv")
dataset = dataset.dropna()


features = dataset.iloc[:,0:13]
label = dataset.iloc[:,13]
```

```python
scaler = StandardScaler()
newDataset = scaler.fit_transform(features)
df = pd.DataFrame(newDataset)
```

```python
plt.boxplot(df)
plt.show()
```

```python
plt.hist(df)
plt.show()
```

```python
sns.heatmap(df.corr(), fmt=".1f",annot=True)
plt.show()
```
